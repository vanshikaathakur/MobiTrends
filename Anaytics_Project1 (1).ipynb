```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_excel(r"C:\Users\nt397\Downloads\Mobiles Dataset (2025).xlsx")
df
```

```python
df.isnull().sum()
```

```python
price_columns=["Launched Price (Pakistan)","Launched Price (India)","Launched Price (China)","Launched Price (USA)","Launched Price (Dubai)"] 
price_columns
```

```python
for i in price_columns:
    df[i] = df[i].replace(
        ['Not available', 'N/A', 'null', 'NaN', 'missing'], 
        0
    )
```

```python
s=df[price_columns]
```

```python
for col in price_columns:
    df[col] = df[col].map(lambda x: ''.join(c for c in str(x) if c.isdigit()))
df
```

```python
df.info()
```

```python
df[price_columns]=df[price_columns].astype('int')
df.info()
```

```python
import matplotlib.pyplot as plt
```

```python
for col in price_columns:
    plt.boxplot(df[col].dropna())
    plt.title(f'Boxplot for {col}')
    plt.ylabel('Price')
    plt.show()
```

```python
s=df['Launched Price (Pakistan)']
```

```python
q1=np.quantile(s,0.25)
q1
```

```python
q2=np.quantile(s,0.75)
q2
```

```python
h=q2-q1
h
```

```python
upper=q2+(1.5*h)
upper
```

```python
lower=q1-(1.5*h)
lower
```

```python
p=[]
for i in s:
    if(i>lower and i<upper):
        p.append(i)
```

```python
plt.boxplot(p)
plt.title(f'Boxplot for Launched Price (Pakistan) ')
plt.show()
```

```python
# Price Analysis by Region
df_avg_prices = df.groupby('Launched Year')[[
    'Launched Price (Pakistan)',
    'Launched Price (India)',
    'Launched Price (China)',
    'Launched Price (USA)',
    'Launched Price (Dubai)'
]].mean()

ax = df_avg_prices.plot(kind='bar', figsize=(12, 8), edgecolor='black')

# Set labels and title
plt.xlabel("Launched Year")
plt.ylabel("Average Launched Price")
plt.title("Average Launched Price by Region and Year")
plt.xticks(rotation=45)

# Add a legend to differentiate the regions
plt.legend(title="Region")

# Adjust layout to prevent labels from overlapping
plt.tight_layout()

```

```python
# Feature vs. Price Analysis
df_feature_price_analysis = df.groupby([
    'RAM',
    'Front Camera',
    'Back Camera',
    'Processor',
    'Battery Capacity',
    'Screen Size'
])[[
    'Launched Price (Pakistan)',
    'Launched Price (India)',
    'Launched Price (China)',
    'Launched Price (USA)',
    'Launched Price (Dubai)']].sum()

print(df_feature_price_analysis.head())
```

```python
import seaborn as sns
features = ['RAM']
price_col = 'Launched Price (India)'
    

for feature in features:
    plt.figure(figsize=(8, 5))
    sns.scatterplot(data=df, x=feature, y=price_col, hue='Launched Year', palette='viridis')
    plt.title(f'{feature} vs {price_col}')
    plt.xlabel(feature)
    plt.ylabel(price_col)
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
```

```python
# Model Analysis
model_prices= df.groupby('Model Name')[[
    'Launched Price (Pakistan)',
    'Launched Price (India)',
    'Launched Price (China)',
    'Launched Price (USA)',
    'Launched Price (Dubai)'
]].mean()
model_prices
```

```python
# Sort by average price in India and take top 10
top_10_models = model_prices.sort_values(by=[
    'Launched Price (Pakistan)',
    'Launched Price (India)',
    'Launched Price (China)',
    'Launched Price (USA)',
    'Launched Price (Dubai)'
], ascending=False).head(10)


plt.figure(figsize=(14, 8))
top_10_models.plot(kind='bar', edgecolor='black', ax=plt.gca())

plt.xlabel("Model Name")
plt.ylabel("Average Launched Price")
plt.title("Top 10 Models by Average Launched Price")
plt.xticks(rotation=45, ha='right')
plt.legend(title="Region")
plt.tight_layout()
plt.show()
```

```python
df['Launched Year'] = df['Launched Year'].astype(str)


df_avg_prices = df.groupby('Launched Year')['Launched Price (India)'].mean()

plt.figure(figsize=(12, 8))
df_avg_prices.plot(kind='bar', edgecolor='black', ax=plt.gca())

plt.xlabel("Launched Year")
plt.ylabel("Average Launched Price")
plt.title("Average Launched Price by Region and Year")
plt.xticks(rotation=45)
plt.legend(title="Region")
plt.tight_layout()
plt.savefig('average_prices_bar_chart.png')  # Save the bar chart
plt.close()


plt.figure(figsize=(10, 6))

sns.violinplot(x='Launched Year', y='Launched Price (India)', data=df, palette='viridis')

plt.title('Distribution of Launched Price (India) by Year')
plt.xlabel('Launched Year')
plt.ylabel('Launched Price (India)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

plt.show()
```
